# Report figures

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

# Baseline algorithm with written feedback

#### Number of step per episodes

```python
mat_step_times_base = []
for i in range(0, 3):
   file = "./data_base/step_record_" + str(i) + ".npy"
   temp = np.load(file)
   mat_step_times_base.append(temp)
mat_step_times_base_somme = np.sum(mat_step_times_base, axis = 0)
mat_step_times_base_somme
```

```python
test1=mat_step_times_base_somme

box_plot_data=[test1]
plt.boxplot(box_plot_data,patch_artist=True,labels=[''])
plt.ylabel("Number of step for 3 episodes")
plt.ylim((0, 60))
plt.savefig("step_total.png")
plt.show()
```

#### Number of rat caught per steps (for 3 episodes)

```python
mat_step_times_base
```

```python
mat_catch_times_base = []
for i in range(0, 3):
   file = "./data_base/catch_times_" + str(i) + ".npy"
   temp = np.load(file)
   mat_catch_times_base.append(temp)

mat_catch_times_base = np.divide(mat_catch_times_base, mat_step_times_base)
mat_catch_times_base = np.mean(mat_catch_times_base, axis = 0)

test1 = mat_catch_times_base

box_plot_data=[test1]
plt.boxplot(box_plot_data,patch_artist=True,labels=[''])
plt.ylabel("Number of rat caught per steps")
plt.ylim((0, 0.5))
plt.savefig("catch_step_results.png")
plt.show()
```

#### Gap between the learned policy and the optimal policy

```python
mat_distance_optimal = []
for i in range(0, 3):
   file = "./data_base/distance_optimal" + str(i) + ".npy"
   temp = np.load(file)
   mat_distance_optimal.append(temp)

mat_distance_optimal = np.mean(mat_distance_optimal, axis = 0)

test1=mat_distance_optimal

box_plot_data=[test1]
plt.boxplot(box_plot_data,patch_artist=True,labels=[''])
plt.ylabel("Gap between the learned and the optimal policy")
plt.ylim((0, 8))
plt.savefig("gap_results_base")
plt.show()
```

# Comparaison with other feedback

#### Catch per state

```python
mat_catch_times_base_nofb = []
for i in range(0, 3):
   file = "./data_base/catch_times_no_feedback_" + str(i) + ".npy"
   temp = np.load(file)
   mat_catch_times_base_nofb.append(temp)
mat_catch_times_base_nofb = np.mean(mat_catch_times_base_nofb, axis = 0)

mat_catch_times_base = []
for i in range(0, 3):
   file = "./data_base/catch_times_" + str(i) + ".npy"
   temp = np.load(file)
   mat_catch_times_base.append(temp)
mat_catch_times_base = np.mean(mat_catch_times_base, axis = 0)

mat_catch_times_gestures = []
for i in range(0, 3):
   file = "./data_gesture/catch_times_" + str(i) + ".npy"
   temp = np.load(file)
   mat_catch_times_gestures.append(temp)
mat_catch_times_gestures = np.mean(mat_catch_times_gestures, axis = 0)

mat_catch_times_son = []
for i in range(0, 3):
   file = "./data_son/catch_times_" + str(i) + ".npy"
   temp = np.load(file)
   mat_catch_times_son.append(temp)
mat_catch_times_son = np.mean(mat_catch_times_son, axis = 0)
```

```python
test1=mat_catch_times_base_nofb
test2=mat_catch_times_base
test3=mat_catch_times_gestures
test4=mat_catch_times_son

box_plot_data=[test1,test2,test3,test4]
plt.boxplot(box_plot_data,patch_artist=True,labels=['No feedback', 'Written','Gesture', 'Son'])
plt.ylabel("Number of rat caught per states")
plt.savefig("catch_state_results_comparaison")
plt.show()
```
